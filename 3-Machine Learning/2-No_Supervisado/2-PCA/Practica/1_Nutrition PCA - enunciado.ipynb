### Distribuciones
Muchas de las variables tienen asimetría hacia la derecha. Deberíamos transformarlas para conseguir distribuciones normales y mejorar las correlaciones de cara al PCA y a futuros modelos lineales que vayamos a probar. Transforma todas las variables realizando una transformación Logarítmica.

```python
plt.hist(df_num['Sugar_g'], bins=30, color='skyblue', edgecolor='black')
plt.title('Sugar_g (Original)')
plt.show()
```

```python
df_log = np.log1p(df_num)
```

```python
plt.hist(df_log['Sugar_g'], bins=30, color='lightgreen', edgecolor='black')
plt.title('Sugar_g (Log1p)')
plt.show()
```

```python
print(f"Skew original: {df_num['Sugar_g'].skew():.4f}")
print(f"Skew log1p:    {df_log['Sugar_g'].skew():.4f}")
```

### Estandarizado
Estandariza cada variable.

No es necesario que dividas en train y test.

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_log)
df_pca = pd.DataFrame(df_scaled, columns=df_log.columns)
df_pca.head()
```

### Implementación del PCA
Prueba a montar un PCA con todos los componentes. Para ello utiliza los datos previamente transformados y escalados.

```python
print("Shape original:", df_pca.shape)

pca = PCA()
X_pca = pca.fit_transform(df_pca)
print("Shape PCA:", X_pca.shape)
```

### Aportación de cada PCA
Visualiza en un diagrama de líneas la suma acumulada de la varianza explicativa del PCA.

Si tuviéses que quedarte con 70-75 % de la varianza original, ¿con cuántos Principal Components te quedarías?

```python
var_acumulada = pca.explained_variance_ratio_.cumsum()
plt.plot(var_acumulada, marker='o')
plt.xlabel("Nº de componentes principales")
plt.ylabel("Ratio de varianza explicada acumulada")
plt.title("Ejercicio PCA")
plt.xticks(np.arange(1, len(df_pca.columns) + 1))
plt.grid(True)
plt.show()


print(var_acumulada)
```

Crea un nuevo PCA con X Principal Components, siendo X la cantidad de PC escogidos en el apartado anterior.

### ¿Qué grupo de comida tiene los valores más altos en cada categoría?
Determina para cada Principal Component, los 3 grupos de comida (*FoodGroup*) con los valores del PC más altos.

```python
df_original = pd.read_csv('data/nndb_flat.csv', index_col='ID')
print(df_original['FoodGroup'].value_counts())


```

```python

df_pca6['FoodGroup'] = df_original['FoodGroup']
df_pca6.head()
```

```python

pc_means = df_pca6.groupby('FoodGroup')[['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6']].mean()

top3_by_pc = {}
for col in ['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6']:
    top3_by_pc[col] = pc_means.sort_values(by=col, ascending=False).head(3).index.tolist()


for pc, groups in top3_by_pc.items():
    print(f"{pc}: {groups}")
```
