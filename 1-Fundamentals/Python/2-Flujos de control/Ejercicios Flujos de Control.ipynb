## Ejercicio 1
Declara una variable numérica que será una hora cualquiera del día. Implementa mediante sentencias `if/elif/else` la siguiente casuística:

* Si es entre las 0 y las 8, print "Durmiendo"
* Si es entre las 9 y las 18, print "Trabajando"
* Si es entre las 19 y las 21, print "Clase"
* Si es entre las 22 y las 24, print "Descanso"
* En cualquier otro caso, print "Transporte o error"

```python
hora = int(float(input("¿Qué hora es?")))
if hora in range(8):
    print("Durmiendo")
elif hora in range(9,18):
    print("Trabajando")
elif hora in range(19,21):
    print("Clase")
elif hora in range(22,24):
    print("Descanso")
else:
    print("Transporte o error")
```

## Ejercicio 2
En este ejercicio vamos a implementar un calculador de precios de casas muy sencillo. Tenemos las siguientes variables:

> superficie
>
> distrito

Implementa mediante sentencias `if/elif/else` la siguiente casuística:

1. Si el distrito es "Moncloa" o "Centro", y además la superficie es superior a 100 metros cuadrados, el precio de la casa es de 1000
2. Si el distrito es "Salamanca", y además la superficie de la casa es al menos de 150 metros, el precio de la casa es de 1500
3. Si el distrito no es "Retiro" y la superficie está entre 60 y 80 metros, el precio es de 600
4. En cualquier otro caso, el precio será de 0

```python
superficie = int(input('Indica la superficie total en m2: '))
distrito = input('Indica el distrito: ').lower()


if distrito == 'moncloa' or distrito == 'centro' and superficie > 100:
    precio = 1000
elif distrito == 'salamanca' and superficie < 150:
    precio = 1500
elif distrito != 'retiro' and superficie > 60 and superficie < 80:
    precio = 600
else:
    precio = 0

print(precio)
```

## Ejercicio 3
En este ejercicio vamos a realizar un programa muy parecido. Para este caso queremos que se cumplan las siguientes condiciones:

1. Primero se compruebe si el distrito es "Retiro". Si es asi, que imprima "Distrito Retiro", y si no, "Otro distrito"
2. Si el distrito es "Retiro", comprueba si si la superficie es mayor de 100 metros cuadrados. En tal caso que imprima un precio de 1000, y si no, de 500.

Hay que usar `ifs` anidados

```python
superficie = int(input('Indica la superficie total en m2: '))
distrito = input('Indica el distrito: ').lower()

if distrito == 'retiro':
    print('Distrito Retiro')
    if superficie > 100:
        precio = 1000
        print(f"El precio es {precio}")
else:
    print('Otro distrito')
    precio = 500
    print(f"El precio es {precio}")
```

## Ejercicio 4
Escribe un programa que tenga dos variables: un numero, y una lista numérica. El programa debe recorrer la lista e imprimir por pantalla cada elemento de la lista multiplicado por el número

```python
numero = 2
lista_num = [1, 2, 3, 4]
lista_num2 = []


for x in lista_num:
    y = x * numero
    lista_num2.append(y)
    print(y)
print(lista_num2)
```

## Ejercicio 6
Dada la siguiente lista:

> list1 = [12, 15, 32, 42, 55, 75, 122, 132, 150, 180, 200]

Implementa un programa que los recorra e imprima por pantalla todos los divisibles por 5. Si nos encontramos con alguno que sea mayor que 150, detener el bucle.

```python
list1 = [12, 15, 32, 42, 55, 75, 122, 132, 150, 180, 200]

for x in list1:
    if x > 150:
        break
    if x % 5 == 0:
        print(x)
    
```

NOTA: NO hay lineas en blanco entre una línea y otra.

## Ejercicio 9
En este ejercicio vamos a crear un pequeño juego. Se trata de intentar adivinar un numero del 1 al 5. Tenemos dos intentos para acertar. Pasos a seguir:

1. Ya viene implementado cómo obtener un número aleatorio del 1 al 5
2. Tendrás que declarar en una variable el numero de vidas, y mediante un bucle while, comprobar que todavia quedan vidas.
3. Dentro del bucle, obtener el valor del usuario y comprobar si es ese el numero a adivinar. Si no, actualizar las vidas.
4. Si acertamos, salimos del bucle e imprimimos por pantalla "You win". Y si perdemos también salimos del bucle, pero en este caso imprimimos por pantalla "You lose".

TIP: te puede resultar útil usar la sentencia `else` cuando acabe el bucle `while`. Lo que haya dentro de ese `else` se ejecutará una vez acabe la ejecución del `while`. Lo podrás usar para cuando pierdas.

```python
from random import randint
random_number = randint(1,5)
hearts = 5
print(random_number)
user = int(input('Dame un número del 1 al 5: '))


while hearts > 0:
    print(f"Te quedan {hearts} vidas")
    if user == random_number:
        print('Has acertado!')
        break
    else:
        print('Has fallado, vuelve a intentarlo!')
        user = int(input('Dame un número del 1 al 5: '))
        hearts -= 1
else:
    print('Te has quedado sin corazones! :(')


```
