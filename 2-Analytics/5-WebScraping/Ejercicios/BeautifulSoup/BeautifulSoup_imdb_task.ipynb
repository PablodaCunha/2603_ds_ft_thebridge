## Web scraping de IMDB

Descarga la información correspondiente y guarda en un csv el top de las 250 películas mediante webscraping. Encapsúlalo en un script.

Obtén:
* Título
* Año
* Duración
* Posición
* Rating

```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
import numpy as np
```

```python
url = "https://books.toscrape.com/"


response = requests.get(url)

response.connection
```

```python
if response.status_code >= 200 and response.status_code < 300:
    print("Successfully fetched the page:", response.status_code)
else:
    print("Error:", response.status_code)
```

```python
html = response.content

print(html)

print(response.status_code)
```

```python
soup = bs(html, "lxml")

print(soup.title)

print(soup.small)

print(soup.h1)
```

```python
# Obtén:
# * Título
# * Año
# * Duración
# * Posición
# * Rating
## Como he cambiado de web porque mo estaba negando los requests, voy a hacer una nueva lista de cosas que obtener :´(
# Obtén 2:
# * Título
# * Precio
# * In stock
# * Category
# * Rating
```

```python
# * Título


books_list = []
# print(soup.contents)

books = soup.find_all("article", class_="product_pod")

books

for book in books:
    book_title = book.find('h3').find('a').get('title')
    books_list.append(book_title)

print(books_list)

```

```python
# * Precio
price_list = []

prices = soup.find_all("div", class_="product_price")
prices

for price in prices:

    price_tag = price.find('p', class_='price_color').text
    price_list.append(price_tag)

print(price_list)
len(price_list)
```

```python
# * In stock
instock_list = []

stocks = soup.find_all("p", class_ = 'instock availability')
stocks

books = soup.find_all("article", class_="product_pod")

for book in books:
    stock = book.find('p', class_='instock availability').text
    clean_stock = ''.join(c for c in stock if c.isalnum())
    instock_list.append(clean_stock)

print(instock_list)

len(instock_list)

    

```

```python
# Ratings

books = soup.find_all("article", class_="product_pod")
rate_list = []
rating_guide = {
    'One': 1,
    'Two': 2,
    'Three': 3,
    'Four': 4,
    'Five': 5
}
for book in books:
    rating = book.find('p', class_='star-rating').get('class')
    num_stars = rating_guide[rating[1]]
    rate_list.append((num_stars))
print(rate_list)
```

```python
# * Category

## Saco la URL específica de cada libro, porque la categoría de cada libro sólo aparece dentro de cada artículo

books = soup.find_all("article", class_="product_pod")
url_list = []

for book in books:
    header = book.find('h3').find('a')
    spec_url = header['href']
    # print(spec_url)
    url_t = url + spec_url.lstrip("/")
    # print(url_t)
    url_list.append(url_t)
print(url_list)


```

```python
# Saco la lista completa de categorías de la página principal


categories = soup.find('ul', class_='nav nav-list')
cat_list = []

if categories is None:
    print("No se encontró la lista de categorías")
else:
    lis = categories.find_all('li')
    for li in lis:
        anchor = li.find('a')
        if anchor:
            cate = anchor.text
            cat_list.append(''.join(c for c in cate if c.isalnum()))
            # print(cate)

cat_list.pop(0)
print(cat_list)
```

```python
# De cada libro consulto la barra de navegación para sacar la categoría en la que aparece
cat_list = []
for url in url_list:
    response = requests.get(url)
    soup_book = bs(response.content, "lxml")

    breadcrumb = soup_book.find('ul', class_='breadcrumb')
    if breadcrumb:
        categories = breadcrumb.find_all('a')
        for cate in categories:
            cat = cate.get_text()
            if cat in ['Home', 'Books']:
                continue
            if cat == 'Default':
                cat = 'Special'
            print(cat)
            cat_list.append(cat)
```

```python
from pprint import pprint
complete_list = []

books_list
price_list
cat_list
instock_list
for i in range(len(books_list)):
    book_data = {"atitle": books_list[i], "category": cat_list[i], "in stock": instock_list[i], "price": price_list[i], "rating": f"{rate_list[i]}/5"}
    complete_list.append(book_data)

pprint(complete_list)
```

```python
df = pd.DataFrame(complete_list)

df
```
