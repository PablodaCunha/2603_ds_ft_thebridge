## Ejercicio 9
Crea una matriz aleatoria de 5x5 y estandariza sus valores. Para normalizar tienes que restar sus valores por la media(`np.mean`) y dividir el resultado por la desviación estándar(`np.std`).

````python
array9 = np.random.randint(1, 9, (5, 5))

array9_mean = array9 - np.mean(array9)

array9_std = array9 / np.std(array9)

print(array9)

print(array9_mean)

print(array9_std)
````

````python
print(np.mean(array9, axis=0))
print(np.mean(array9, axis=1))
print(np.mean(array9, None))
````

## Ejercicio 10
1. Crea una secuencia de números que vayan del 0 al 10.
2. Crea un nuevo array, que parta del anterior, pero sustituya todos los elementos entre el 3 y el 8, por el valor -1
3. Crea un nuevo array que niegue todos los elementos entre el 3 y el 8

````python
array10 = np.array(range(11))



array10[3:9] = -1


print(array10)

# array10[3:9] = -array10[3:9] 

# print(array10)

````

````python
array10 = np.array(range(11))



array10[3:9] = -array10[3:9] 

print(array10)
````

````python
array10 = np.array(range(11))


print(np.where((array10 > 3) & (array10 < 8), -1, array10)) # apartado 2
````

````python
array10[(array10 > 3) & (array10 < 9)] = -1

array10
````

````python
array10 = np.array(range(11))
array10[(array10 > 3) & (array10 < 9)] *= -1

array10
````
