# Ejercicio 3 
Feature Engineering con Regex, Str Ops, Numpy
- extrae las distancias y conviértelas a unidades con sentido

1.3.1. Crea una columna llamada distancias
- usa regex para extraer los números
- usa .astype() para convertirlos en el tipo adecuado

1.3.2. Crea una columna llamada unidades
- usa regex para extraer las unidades
- entiende los valores que aparecen en la columna unidad

1.3.3. Convierte las millas a kilómetros


````python
#run this cell to create the dataset
df_distances = pd.DataFrame({
    "city1": ["Madrid", "Madrid", "Madrid", "London", "London", "London"],
    "city2": ["Barcelona", "Valencia", "Sevilla", "Manchester", "Liverpool", "Birmingham"],
    "distance": ["505 kilometres", "303KM", "390.5 km", "163 miles", "220 mi.", "miles: 120"]
})
````

````python
df_distances
````

````python
df_distances['distances_num'] = df_distances['distance'].str.extract(r'(\d+\.?\d*)')[0].astype(float)
````

````python
df_distances['distances_num']
````

````python
df_distances['distances_raw']=df_distances.distance.replace(to_replace=r"kilometres|KM|km|miles|:|mi\.", value="", regex=True)
````

````python
df_distances = df_distances.drop('distances_raw', axis=1)
````

````python
df_distances
````

````python
df_distances['units'] = df_distances['distance'].str.extract(r'([a-zA-Z]+)')[0].str.lower()
````

````python
df_distances['units']
````

````python
is_mile = df_distances['units'].isin(['miles', 'mi'])
````

````python
df_distances['num2'] = df_distances['distances_num'].copy()
````

````python
df_distances.loc[is_mile, 'distances_num'] = df_distances.loc[is_mile, 'distances_num'] * 1.609
````

````python
df_distances['in_Km']=df_distances['distances_num'].copy()
````

````python
df_distances
````

````python
df_distances["in_Km2"] = 0

def to_km(row):
    #print(row)
    if "k" in row.units or "K" in row.units:
        return row.num2
    else:
        return row.num2 * 1.609  # 1 mile is equal to 1.609 km
    
df_distances["in_Km2"] = df_distances.apply(to_km, axis=1)
````
