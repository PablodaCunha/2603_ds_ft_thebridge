```python
# 19. Veamos la función melt() de pandas. Melt toma los valores de varias columnas y los junta todos en una misma columna nueva.
# Dado un DataFrame, melt recibe varios parámetros:
# * qué columnas del DataFrame se quedan igual
# * qué nombres de columnas del dataframe pasan a ser una columna nueva
# * cómo queremos llamar esas nuevas columnas

# Primero recuerda el Dataframe df_grass_fire_water con las primeras 5 filas
```

```python
df_grass_fire_water.head(5)
```

```python
df_2 = pd.melt(df_grass_fire_water, id_vars=['Name', 'Type 1', 'Type 2'], value_vars=['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed'], var_name='Power', value_name='Stat')

df_2.head()
```

```python
df_3 = df_grass_fire_water.melt(id_vars=['Name', 'Type 1'], value_vars=['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed'], var_name='New Column')
```

```python
# 26. Usa catplot con el DataFrame original. Muestra el valor de Defense según el 'Type 1', muestra un subplot para legendarios y otro para no legendarios
# y divide los colores según el Type 2
```

```python
df = pd.read_csv('Pokemon.csv', index_col=0)
```

```python
f1 = sns.catplot(
    data=df,
    x='Type 1', 
    y='Defense', 
    col='Legendary',
    hue='Type 2',     
    kind='box',       
    height=5, aspect=1, 
)
for ax in f1.axes.flat:
    ax.tick_params(axis='x', rotation=90)
plt.suptitle('Defense por Type 1 y Legendary (coloreado por Type 2)')
plt.tight_layout()
plt.show()
```

```python
f2 = sns.catplot(data=df, x='Type 1', y='Defense', col='Legendary', hue='Type 2', kind='violin', inner="point")
f2.tick_params(axis='x', rotation=90);
```

```python
# 27. Muestra una función distribución de probabilidad con kdeplot (curvas de nivel) de la defensa respecto al ataque.
# Di un bin 2D donde caigan muchos puntos (ataque, defensa).
```

```python
plt.figure(figsize=(10,8))
sns.kdeplot(data=df, x='Attack', y='Defense', fill=True, cmap='Blues', levels=10)
plt.title('Distribucion ataque y defensa')
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.axhline(y=df['Defense'].mean(), color='red', linestyle='--', label='Defense')
plt.axvline(x=df['Attack'].mean(), color='orange', linestyle='--', label='Attack')
plt.legend();
```

```python
# 29. Pinta una distribución de probabilidad conjunta, pintando la defensa respecto al ataque del DataFrame original
```

```python
sns.jointplot(
    data=df, 
    x='Attack', 
    y='Defense',
    kind='hex',  
    bins=50,    
    height=10
)
plt.suptitle('Distribución Conjunta');
```
